```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context('notebook', font_scale=1.5)
sns.set_style('white')
import matplotlib.pylab as plt
%matplotlib inline
import os
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
ff_data_dir = '~/FragileFamiliesChallenge/FFChallenge/'

bg_f = os.path.join(ff_data_dir, 'background.csv')
train_f = os.path.join(ff_data_dir, 'train.csv')
```

```python
df_bg = pd.read_csv(bg_f, low_memory=False)
vartype_df = pd.read_csv("../helper_data/ffc_variable_types.csv")
```

```python
varcont = vartype_df["variable"][vartype_df["variable_type"] == "continuous"]
varcat = vartype_df["variable"][vartype_df["variable_type"] == "categorical"]
dfcont = df_bg[varcont]
dfcat = df_bg[varcat]
```

```python
#index: var name
# col 1: how many nan values
# col 2 how many unique values
# col last: description
dfcat_describe = pd.DataFrame(index=dfcat.columns)
dfcat_describe['NaNs'] = [dfcat[col].isnull().sum() for col in dfcat.columns]
dfcat_describe['unique values'] = [dfcat[col].nunique() for col in dfcat.columns]
dfcat_describe['decsription'] = [vartype_df.loc[vartype_df['variable'] == col]['label'] for col in dfcat.columns]
```

```python
# The 12 variable with the most unique values appear not to be categorical. I will change the csv
dfcat_describe.sort_values(by='unique values', ascending=False)
#csv now changed
```

```python
new_vartype_df=vartype_df
for var in dfcat_describe.sort_values(by='unique values', ascending=False).head(12).index:
    new_vartype_df.loc[new_vartype_df["variable"]==var,"variable_type"] = "continuous"
```

```python
new_vartype_df.to_csv("../helper_data/ffc_variable_types.csv")
```
